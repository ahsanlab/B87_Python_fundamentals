```python
# Write Python code to find average via loop

def cal_average(num):
    sum_num =0
    for t in num:
        sum_num=sum_num+t

    avg = sum_num / len(num)
    return avg
print("The average is", cal_average([18,25,3,41,5]))
```

```python
#Code for to get index of an element in a list using for loop

my_list = ['Guru', 'Ahsan', 'Siya', 'Tiya', 'Daksh', 'Riya', 'Guru']
all_index = []
for i in range(0, len(my_list)):
    if my_list[i] == 'Guru':
        all_index.append(i)
print("OriginalList", my_list)
print("Index for element Guru : ", all_index)
```
